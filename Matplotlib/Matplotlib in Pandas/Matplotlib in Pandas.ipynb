```python
Over50 = X[X["age"]>50]
Over50.head(5)
```

```python
fig, ax = plt.subplots()
ax.scatter(x=Over50.age, y=Over50.chol,c=Over50.target,label=Over50.target)
ax.set(title="Chol Vs Age",xlabel="Age",ylabel="Chol")
ax.set_xlim(50,90)
ax.axhline(Over50.chol.mean(),ls='--')
```

```python
plt.style.use('seaborn-v0_8-whitegrid')

fig, ((ax0,ax1)) = plt.subplots(nrows=2,ncols=1,figsize = (10,10),sharex=True)

scatter = ax0.scatter(x = Over50.age, y = Over50.chol, c = Over50.target,
                       cmap="winter") # Changes Color Schema
ax0.set(title="Heart Disease and Cholesterol Levels",ylabel="Cholesterol")
ax0.legend(*scatter.legend_elements(),title="Target");
ax0.axhline(Over50.chol.mean(),ls='--')


scatter1 = ax1.scatter(x = Over50.age, y = Over50.thalach, c = Over50.target,
                        cmap="winter")
ax1.set(title="Heart Disease and Max Heart Rate", xlabel="Age", ylabel="Max Heart Rate")
ax1.legend(*scatter1.legend_elements(),title="Target1")
ax1.axhline(Over50.thalach.mean(), ls='--')

fig.suptitle("Heart Disease Analysis",fontsize=15,fontweight = "bold")
```
