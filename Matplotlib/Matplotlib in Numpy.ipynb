# MATPLOTLIB

### Import Matplotlib

```python
%matplotlib inline
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

a1 = np.arange(10)
a2 = np.arange(10)

fig, ax = plt.subplots(figsize = (5,5))

ax.plot(a1,a2)

ax.set(title = "simple plot",xlabel = "X axis", ylabel = "Y axis")
```

```python
fig.savefig("Matplotlib/Sample-plot.png")
```

## Making Figures With Numpy

```python
x = np.linspace(0,100,20)
x
```

### Line Plot

```python
fig, ax = plt.subplots()
ax.plot(x,x**2)
```

### Scatter Plot

```python
fig, ax = plt.subplots()
ax.scatter(x, x**2)
```

```python
x1 = np.linspace(0,10,100)
fig, ax = plt.subplots()
ax.scatter(x1, np.sin(x1));
```

# Bar Plot

### Make A plot from Dictionary

```python
prices = {"Nike" : 40,"Adidas":35,"New Balance": 30 }
```

```python
fig, ax = plt.subplots()
ax.bar( list(prices.keys()), list(prices.values()))
ax.set(title="Sneakers Sold this month", ylabel="Items",xlabel="Brands")
```

## Horizontal Bar Plot

```python
fig, ax = plt.subplots()
ax.barh(list(prices.keys()),list(prices.values()),height=0.8)
```

# Histograph

```python
x2 = np.random.randn(1000)
x2[:10]
```

```python
fig, ax = plt.subplots()
ax.hist(x2)
```

# Subplots

```python
fig, ((ax1,ax2),(ax3,ax4)) = plt.subplots( nrows=2,
                                           ncols=2,
                                        figsize=(10,5))

ax1.plot(x,x*2)
ax2.scatter(x1,np.sin(x1))
ax3.bar(list(prices.keys()),list(prices.values()))
ax4.hist(x2)
```

```python
fig.savefig("Matplotlib.subplots.png")
```
